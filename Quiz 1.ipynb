### Range

```python
num = (1, 2, 3, 4, 5)
print(num)

for x in range(num[0], num[-1]+1):
    print(x)

data = range(0, 11)
print(list(data))
```

### Event = Kelipatan

```python
event = range(2, 21, 2)
print(list(event))
```

### Data Frame

```python
import pandas as pd
# Data frame dari dictionary
data_start = {
    "Nama": ["Marsha", "Budi", "Siti", "Ani"],
    "Umur": [17, 21, 19, 22],
    "Kota": ["Waingapu", "Bandung", "Surabaya", "Medan"],
    "Status": ["Mahasiswa", "Mahasiswa", "Mahasiswa", "Mahasiswa"]
}

df = pd.DataFrame(data_start)
print(df)
```

```python
# Data frame dari list of lists
data_finish = [
    ["Marsha", 17, "Waingapu", "Mahasiswa"],
    ["Budi", 21, "Bandung", "Mahasiswa"],
    ["Siti", 19, "Surabaya", "Mahasiswa"],
    ["Ani", 22, "Medan", "Mahasiswa"]
]
df2 = pd.DataFrame(data_finish, columns=["Nama", "Umur", "Kota", "Status"])
print(df2)
print(" ")
print(df2.columns.to_list())
print(" ")
df2.columns = ["Nama", "Umur", "Kota", "Yehey"]
print(df2)
```

```python
print(df2[["Nama", "Umur"]])
```
