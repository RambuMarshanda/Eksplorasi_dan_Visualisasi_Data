## Deskripsi Data

Data ini mencakup informasi mengenai **Rata-Rata Lama Sekolah (RLS)** dan **Tingkat Pengangguran Terbuka (TPT)** di berbagai provinsi di Indonesia. RLS menunjukkan rata-rata lama pendidikan yang ditempuh oleh penduduk berusia 15 tahun ke atas, sedangkan TPT mencerminkan persentase pengangguran di setiap provinsi. Data ini memberikan gambaran mengenai hubungan antara tingkat pendidikan dan pengangguran, serta perbedaan antarprovinsi terkait kondisi ketenagakerjaan dan pendidikan, yang dapat digunakan untuk analisis sosial dan perencanaan kebijakan yang lebih tepat.


## Import & Preview Data

```python
import pandas as pd

data = pd.read_excel("RLS dan TPT.xlsx")

print(data)

sorted_data = data.sort_values(by='Rata-Rata Lama Sekolah Penduduk Umur 15 Tahun ke Atas Menurut Provinsi')
sorted_data = data.sort_values(by='Tingkat Pengangguran Terbuka Menurut Provinsi (Persen)')
sorted_data = data.sort_values(by='38 Provinsi')
```

```python
# Visualisasi Histogram

import matplotlib.pyplot as plt
plt.figure(figsize=(8, 6))

plt.hist(sorted_data['Rata-Rata Lama Sekolah Penduduk Umur 15 Tahun ke Atas Menurut Provinsi'], bins=5, color='green', edgecolor='black')

plt.title('Distribusi RLS 2024')

plt.xlabel('RLS')

plt.ylabel('Frekuensi')

plt.show()
```

```python
# Visualisasi Scatter Plot
import pandas as pd
df = pd.read_excel('RLS dan TPT.xlsx')

plt.figure(figsize=(8, 6))
plt.scatter(df['Rata-Rata Lama Sekolah Penduduk Umur 15 Tahun ke Atas Menurut Provinsi'], df['Tingkat Pengangguran Terbuka Menurut Provinsi (Persen)'], color='blue', marker='o')

# Menambahkan label dan judul
plt.title('Hubungan antara RLS (Rata-Rata Lama Sekolah) dan TPT (Tingkat Pengangguran Terbuka)', fontsize=14)
plt.xlabel('Rata-Rata Lama Sekolah (RLS)', fontsize=12)
plt.ylabel('Tingkat Pengangguran Terbuka (TPT)', fontsize=12)

plt.tight_layout()
plt.show()
```

```python
# Visualisasi Pie Chart
import matplotlib.pyplot as plt

# Rata-rata data RLS dan TPT
rata_rata_rls = 9.22  
rata_rata_tpt = 4.82  

# Data untuk pie chart
categories = ['Rata-Rata Lama Sekolah Penduduk Umur 15 Tahun ke Atas Menurut Provinsi', 'Tingkat Pengangguran Terbuka Menurut Provinsi (Persen)']
values = [rata_rata_rls, rata_rata_tpt]

# Membuat pie chart
plt.figure(figsize=(8, 8))
plt.pie(values, labels=categories, autopct='%1.1f%%', startangle=90, colors=['skyblue', 'salmon'])
plt.title('Perbandingan Rata-Rata Lama Sekolah (RLS) dan Tingkat Pengangguran Terbuka (TPT)')

# Menampilkan chart
plt.show()
```
