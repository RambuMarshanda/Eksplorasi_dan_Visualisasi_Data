## Deskripsi Data

Data ini mencakup informasi mengenai **Rata-Rata Lama Sekolah (RLS)** dan **Tingkat Pengangguran Terbuka (TPT)** di berbagai provinsi di Indonesia. RLS menunjukkan rata-rata lama pendidikan yang ditempuh oleh penduduk berusia 15 tahun ke atas, sedangkan TPT mencerminkan persentase pengangguran di setiap provinsi. Data ini memberikan gambaran mengenai hubungan antara tingkat pendidikan dan pengangguran, serta perbedaan antarprovinsi terkait kondisi ketenagakerjaan dan pendidikan, yang dapat digunakan untuk analisis sosial dan perencanaan kebijakan yang lebih tepat.


## Import & Preview Data

```python
import pandas as pd

data = pd.read_excel("RLS dan TPT.xlsx")

print(data)

sorted_data = data.sort_values(by='Rata-Rata Lama Sekolah Penduduk Umur 15 Tahun ke Atas Menurut Provinsi')
sorted_data = data.sort_values(by='Tingkat Pengangguran Terbuka Menurut Provinsi (Persen)')
sorted_data = data.sort_values(by='38 Provinsi')
```

## Visualisasi Data

```python
# Visualisasi Histogram

import matplotlib.pyplot as plt
plt.figure(figsize=(8, 6))

plt.hist(sorted_data['Rata-Rata Lama Sekolah Penduduk Umur 15 Tahun ke Atas Menurut Provinsi'], bins=5, color='green', edgecolor='black')

plt.title('Distribusi RLS 2024')

plt.xlabel('RLS')

plt.ylabel('Frekuensi')

plt.show()
```

```python
# Visualisasi Scatter Plot

plt.figure(figsize=(8, 6))

plt.scatter(sorted_data['Rata-Rata Lama Sekolah Penduduk Umur 15 Tahun ke Atas Menurut Provinsi'], sorted_data['Tingkat Pengangguran Terbuka Menurut Provinsi (Persen)'], color='red', s=100)

plt.show()
```
