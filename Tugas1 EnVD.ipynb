### Deskripsi Data

Data tersebut menampilkan rata-rata lama sekolah penduduk usia 15 tahun ke atas di 38 provinsi di Indonesia dalam satuan tahun. Angka ini menunjukkan rata-rata jumlah tahun yang telah ditempuh penduduk dalam pendidikan formal, sehingga dapat menggambarkan tingkat pencapaian pendidikan di masing-masing provinsi. Terlihat adanya variasi antarprovinsi, dengan beberapa daerah memiliki rata-rata lama sekolah di atas 10 tahun, sementara sebagian lainnya masih berada di bawah 9 tahun, serta rata-rata nasional Indonesia sebesar 9,22 tahun.

### Split Kolom Data Menggunakan Iloc

```python
import pandas as pd

df = pd.read_excel("RLS.xlsx")

df.head(39)
```

```python
# iloc ke 1 → Kolom Provinsi
iloc_1 = df.iloc[:, 0:1]

# iloc ke 2 → Kolom Rata_Rata
iloc_2 = df.iloc[:, 1:2]
```

```python
iloc_1.head(39)
```

```python
iloc_2.index = iloc_1.index
iloc_2.head(39)
```

### Merge data berdasarkan Index

```python
df_merge = pd.merge(
    iloc_1,
    iloc_2,
    left_index=True,
    right_index=True
)

df_merge.head(39)
```

### Export Output

```python
df_merge.to_csv("hasil_merge_RLS.csv", index=True)
```
