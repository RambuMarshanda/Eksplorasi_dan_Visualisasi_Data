## Deskripsi Data

Data ini mencakup informasi mengenai Rata-Rata Lama Sekolah (RLS) dan Tingkat Pengangguran Terbuka (TPT) di berbagai provinsi di Indonesia. RLS menunjukkan rata-rata lama pendidikan yang ditempuh oleh penduduk berusia 15 tahun ke atas, sedangkan TPT mencerminkan persentase pengangguran di setiap provinsi. Data ini memberikan gambaran mengenai hubungan antara tingkat pendidikan dan pengangguran, serta perbedaan antarprovinsi terkait kondisi ketenagakerjaan dan pendidikan, yang dapat digunakan untuk analisis sosial dan perencanaan kebijakan yang lebih tepat.

## Import & Preview Data

```python
import pandas as pd

df = pd.read_excel("RLS dan TPT.xlsx")
print(df)
```

## Analisis Deskriptif

```python
print(df.columns)
```

```python
import numpy as np
from scipy import stats

data = df['Rata-Rata Lama Sekolah Penduduk Umur 15 Tahun ke Atas Menurut Provinsi'].to_numpy()

mean = np.mean(data)
median = np.median(data)
mode = stats.mode(data)
variance = np.var(data)
std_dev = np.std(data)

print(f"Mean: {mean}")
print(f"Median: {median}")
print(f"Mode: {mode}")
print(f"Variance: {variance}")
print(f"Standard Deviation: {std_dev}")
```

```python
import numpy as np
from scipy import stats

data = df['Tingkat Pengangguran Terbuka Menurut Provinsi (Persen)'].to_numpy()

mean = np.mean(data)
median = np.median(data)
mode = stats.mode(data)
variance = np.var(data)
std_dev = np.std(data)

print(f"Mean: {mean}")
print(f"Median: {median}")
print(f"Mode: {mode}")
print(f"Variance: {variance}")
print(f"Standard Deviation: {std_dev}")
```

## Distribusi Data

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats

rls = df['Rata-Rata Lama Sekolah Penduduk Umur 15 Tahun ke Atas Menurut Provinsi']
tpt = df['Tingkat Pengangguran Terbuka Menurut Provinsi (Persen)']
```

```python
# Histogram
plt.hist(rls, bins=10)
plt.title("Distribusi Rata-Rata Lama Sekolah (RLS)")
plt.xlabel("RLS")
plt.ylabel("Frekuensi")
plt.show()

plt.hist(tpt, bins=10)
plt.title("Distribusi Tingkat Pengangguran Terbuka (TPT)")
plt.xlabel("TPT (%)")
plt.ylabel("Frekuensi")
plt.show()
```

```python
# Boxplot
plt.boxplot(rls)
plt.title("Boxplot RLS")
plt.show()

plt.boxplot(tpt)
plt.title("Boxplot TPT")
plt.show()
```

```python
# Uji Normalitas(Shapiro-Wilk)
print("Uji Normalitas RLS")
stat, p = stats.shapiro(rls)
print("Statistic:", stat)
print("p-value:", p)
if p > 0.05:
    print("Data kemungkinan berdistribusi normal")
else:
    print("Data tidak berdistribusi normal")
```

```python
print("Uji Normalitas TPT")
stat, p = stats.shapiro(tpt)
print("Statistic:", stat)
print("p-value:", p)
if p > 0.05:
    print("Data kemungkinan berdistribusi normal")
else:
    print("Data tidak berdistribusi normal")
```
